# Salary column in integer

```python
df['Salary Estimate'].unique()
```

```python
import re
# Define a function to extract salary range and calculate average
def extract_and_calculate_average(text):
    pattern = r'\$(\d+)K-\$(\d+)K'
    match = re.search(pattern, text)
    
    if match:
        low_salary = int(match.group(1))  # Convert '137' to int
        high_salary = int(match.group(2))  # Convert '171' to int
        average_salary = (low_salary + high_salary) // 2  # Use // for integer division
        return pd.Series({'Low Salary': low_salary, 'High Salary': high_salary, 'Average Salary': average_salary})
    else:
        return pd.Series({'Low Salary': None, 'High Salary': None, 'Average Salary': None})


# Apply the function to the 'salary_text' column
df[['Low Salary', 'High Salary', 'Average Salary']] = df['Salary Estimate'].apply(extract_and_calculate_average)

```

# how to take infomation of columns Job Description

```python
df['python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['hadoop'] = df['Job Description'].apply(lambda x: 1 if 'hadoop' in x.lower() else 0)
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['tableau'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df['big_data'] = df['Job Description'].apply(lambda x: 1 if 'big data' in x.lower() else 0)
df.head()
```

# Remove the number of column Company Name 

```python
#df['Company Name'].unique()
df['Company Name'] = df['Company Name'].str.split('\n').str[0]
```

```python
df['seniority'] = df['Job Title'].apply(lambda title: 'Senior' if any(substring in title.lower() for substring in ['sr', 'senior', 'lead', 'principal']) \
                                        else ('Junior' if any(substring in title.lower() for substring in ['jr', 'jr.']) else 'NA'))
```

```python
df['job_simp'] = df['Job Title'].apply(lambda title: 'Data Scientist' if 'data scientist' in title.lower()\
                                       else ('Data Engineer' if 'data engineer' in title.lower()\
                                             else ('Analyst' if 'analyst' in title.lower()\
                                                   else ('MLE' if 'machine learning' in title.lower()\
                                                         else ('Manager' if 'manager' in title.lower()\
                                                               else ('Director' if 'director' in title.lower() else 'NA'))))))
```
